```python
import pandas as pd
df = pd.read_csv('../data/jobs.csv')
print(df)
```

```python
df['text']= df["job_title"]+" " +df["skills_required"]
df[["job_title","text"]]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
job_vectors = vectorizer.fit_transform(df["text"])
job_vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
def recommend_career(user_input): 
    user_vector= vectorizer.transform([user_input])
    similarity_scores=cosine_similarity(user_vector,job_vectors)
    best_match_index=similarity_scores.argmax()

    return df.iloc[best_match_index]["job_title"]
    
```

```python
recommend_career("Python SQL data analysis")
```

```python
recommend_career("network security linux")
```

```python
def skill_gap_analysis(user_skills, job_index):
    # Normalize user skills
    user_skills = user_skills.replace(",", " ").lower()
    user_skills_set = set(user_skills.split())

    # Normalize job skills
    job_skills = df.iloc[job_index]["skills_required"].lower()
    job_skills_set = set(job_skills.replace(",", " ").split())

    # Find missing skills
    missing_skills = job_skills_set - user_skills_set
    return list(missing_skills)

```

```python
def recommend_with_skills(user_input):
    user_vector = vectorizer.transform([user_input])
    similarity_scores = cosine_similarity(user_vector, job_vectors)
    best_match_index = similarity_scores.argmax()
    
    career = df.iloc[best_match_index]["job_title"]
    missing_skills = skill_gap_analysis(user_input, best_match_index)
    
    return career, missing_skills
```

```python
recommend_with_skills("python, SQL")
```
